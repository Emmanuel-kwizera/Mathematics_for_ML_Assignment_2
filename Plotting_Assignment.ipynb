```python
# Plotting histogram for co2
plt.figure(figsize=(10,2))
plt.hist(df['co2'], bins=20, color='darkred')
plt.title('Distribution of Carbon Dioxide')
plt.show()
```

```python
# The histogram displays the distribution of CO2 emissions for various vehicles.
# Most vehicles emit between 250 and 400 grams of CO2 per mile, indicating a concentration of moderately efficient vehicles.
# The left side of the graph shows fewer vehicles with very low CO2 emissions — likely hybrid or electric models.
# The right tail has fewer entries as well, showing that very high-emission vehicles are less common in the dataset.
# This distribution is slightly right-skewed, which implies there are more low- to mid-emission vehicles than high-emission ones.
```

```python
# Plotting histogram for highway
plt.figure(figsize=(10,2))
plt.hist(df['highway'], bins=25, color='darkred')
plt.title('Histogram of highway column')
plt.show()
```

```python
# The histogram shows the distribution of highway fuel efficiency (MPG) among vehicles.
# Most vehicles have highway mileage between 25 and 40 MPG, with the highest concentration around 30–35 MPG.
# Very few vehicles exceed 50 MPG, suggesting that highly fuel-efficient vehicles (e.g., hybrids) are less common.
# The distribution is slightly right-skewed, with fewer vehicles having extremely high fuel efficiency.
```

```python
# Plotting histogram for ghgScore
plt.figure(figsize=(10,2))
plt.hist(df['ghgScore'], bins=20, color='darkred')
plt.title('Distribution of ghgScore')
plt.show()
```

```python
# This histogram shows the distribution of GHG Scores across the vehicle dataset.
# Most vehicles have a score between 4 and 7, indicating moderate emissions.
# Very few vehicles have scores of 1–2 (very high emissions) or 9–10 (very low emissions).
# The slight central peak suggests a concentration around average environmental performance.
```

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# The heatmap displays the correlation between numerical features in the dataset, helping us understand how they relate to the ghgScore.
# A strong positive correlation (direct proportion, close to +1) means that as one variable increases, the ghgScore also increases.
# A strong negative correlation (inverse proportion, close to -1) means that as one variable increases, the ghgScore decreases.
# These relationships are useful in AI modeling, as they help identify features that significantly affect the target.

# From the heatmap and correlation values:
# - Features like 'feScore', 'comb', 'city', 'highway', 'UCity', and 'UHighway' show a strong direct proportion to ghgScore.
# - Features like 'co2', 'cylinders', and 'displ' show a strong inverse proportion to ghgScore.
# These should be retained for predictive modeling.

# On the other hand, features like 'id', 'year', 'pv2', and 'pv4' show very weak or no correlation with ghgScore.
# These features can be dropped as they provide minimal or no predictive value and may add noise to the model.

# Therefore, we would consider dropping the following columns:
# - 'id' (not meaningful for prediction)
# - 'year' (weak correlation)
# - 'pv2' and 'pv4' (low and inconsistent correlation)
```
