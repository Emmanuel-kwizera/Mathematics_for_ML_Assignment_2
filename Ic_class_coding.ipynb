```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from google.colab import files
uploaded = files.upload()
```

```python
# Load the dataset
df = pd.read_csv("fuel_econ.csv")

# Display 6 rows (4 < rows < 11)
df.head(6)
```

```python
# Plotting histogram for fuel efficiency
plt.figure(figsize=(10,2))
plt.hist(df['co2'], bins=20, color='darkred')
plt.title('Distribution of Fuel Efficiency Based on CO2')
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.hist(df['highway'], bins=20, color='darkred')
plt.title('Distribution of Fuel Efficiency Based on highway')
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.hist(df['ghgScore'], bins=20, color='darkred')
plt.title('Distribution of Fuel Efficiency Based on ghgScore')
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Correlation Heatmap Between Numerical Features')
plt.show()
```
